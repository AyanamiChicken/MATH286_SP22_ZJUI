Here is the code to plot multiple Graphs for IVP1

```python
# Ploting Vector Field for y'
import matplotlib.pyplot as plt
import numpy as np
X,Y = np.meshgrid(np.arange(-3,3,0.25),np.arange(-3,3,0.25))
V = (Y**2+X**2-X)/np.sqrt((Y**2+X**2-X)**2+1**2)
U = 1/np.sqrt((Y**2+X**2-X)**2+1**2)
#Set up a figure
plt.figure(figsize = (6,6))
plt.quiver(X,Y,U,V,units = "width",pivot="mid")
plt.savefig("./plot/plot_vector_ODE1.png")
```

```python
# Ploting Flow for y'
import matplotlib.pyplot as plt
import numpy as np
X,Y = np.meshgrid(np.arange(-3,3,0.25),np.arange(-3,3,0.25))
V = (Y**2+X**2-X)/np.sqrt((Y**2+X**2-X)**2+1**2)
U = 1/np.sqrt((Y**2+X**2-X)**2+1**2)
#Set up a figure
plt.figure(figsize = (6,6))
plt.streamplot(X,Y,U,V)
plt.savefig("./plot/plot_flow_ODE1.png")
```

```python
# Ploting Countour for y'
plt.rcParams['xtick.direction'] = 'in' 
plt.rcParams['ytick.direction'] = 'in' 
stepx = 0.01; stepy = 0.01 
x = np.arange(-10,10,stepx); y = np.arange(-10,10,stepy) 
X,Y = np.meshgrid(x,y) 
def f(X,Y):           
    return((Y**2+X**2-X))
Z = f(X, Y)
fig, ax = plt.subplots(figsize=(8,8),dpi=100)
CS = ax.contourf(X, Y, Z)
CS = ax.contour(X, Y, Z)
ax.set_xlabel('$x$'); ax.set_ylabel('$y$')
plt.xticks(np.arange(-10,11,2))
plt.yticks(np.arange(-10,11,2))
plt.savefig("./plot/plot_contour_ODE1.png")
```

```python
# Ploting 3D field for y'
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
 
# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = (Y**2+X**2-X)
 
# Plot the surface.
surf = ax.plot_surface(X, Y, R, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
 
# Customize the z axis.
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
 
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.savefig("./plot/plot_3D_ODE1.png")
```

Here is the code to plot multiple Graphs for IVP2

```python
# Ploting Vector Field for y'
import matplotlib.pyplot as plt
import numpy as np
X,Y = np.meshgrid(np.arange(-3,3,0.25),np.arange(-3,3,0.25))
V = (Y**2-X**2)*(Y-X)/np.sqrt(((Y**2-X**2)*(Y-X))**2+1**2)
U = 1/np.sqrt(((Y**2-X**2)*(Y-X))**2+1**2)
#Set up a figure
plt.figure(figsize = (6,6))
plt.quiver(X,Y,U,V,units = "width",pivot="mid")
plt.savefig("./plot/plot_vector_ODE2.png")
```

```python
# Ploting Flow for y'
import matplotlib.pyplot as plt
import numpy as np
X,Y = np.meshgrid(np.arange(-10,10,0.25),np.arange(-10,10,0.25))
V = (Y**2-X**2)*(Y-X)/np.sqrt(((Y**2-X**2)*(Y-X))**2+1**2)
U = 1/np.sqrt(((Y**2-X**2)*(Y-X))**2+1**2)
#Set up a figure
plt.figure(figsize = (6,6))
plt.streamplot(X,Y,U,V)
plt.savefig("./plot/plot_flow_ODE2.png")
```

```python
# Ploting Countour for y'
plt.rcParams['xtick.direction'] = 'in' 
plt.rcParams['ytick.direction'] = 'in' 
stepx = 0.01; stepy = 0.01 
x = np.arange(-10,10,stepx); y = np.arange(-10,10,stepy)
X,Y = np.meshgrid(x,y)
def f(X,Y):           
    return((Y**2-X**2)*(Y-X))
Z = f(X, Y)
fig, ax = plt.subplots(figsize=(8,8),dpi=100)
CS = ax.contourf(X, Y, Z)
CS = ax.contour(X, Y, Z)
ax.set_xlabel('$x$'); ax.set_ylabel('$y$')
plt.xticks(np.arange(-10,11,2))
plt.yticks(np.arange(-10,11,2))
plt.savefig("./plot/plot_contour_ODE2.png")
```
